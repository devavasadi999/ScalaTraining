### Iterative Statements

* For loop
* While loop
* Do While loop
* Collection Iteration
* Functional Iteration
* Range Based Looping
* Recursive Iteration
* For loop over Collection
* For with Option and Either

```scala
// Basic for loop

for(i <- 1 to 3) {
    print(s"$i ")
}
println()

// for loop with filters

for(i <- 1 to 3 if i > 1) {
    println(s"$i ")
}
println()

// for loop with multiple generators

for(i <- 1 to 2; j <- 1 to 3) {
    println(s"$i, $j")
}
println()

// for loop with yield

val doubledNums = for(num <- 1 to 3) yield num * 2
println(doubledNums)

// while loop

var i = 0
while (i <= 3) {
    print(s"$i ")
    i+=1
}
println()

// do while loop

var j = 0
do {
    print(s"$j ")
    j += 1
} while (j <= 2)
println()

// collection iteration using for each

val list = List(1, 2, 3)
list.foreach(num => print(s"$num "))
println()

// functional iteration methods

println(list.map(_ * 3))
println(list.filter(_ % 3 != 0))
val nestedList = List(List(1, 2), List(3, 4))
println(nestedList.flatMap(identity))

// Range based looping ( to, until )

for(i <- 1 until 3)
    print(s"$i ")
println()

// Recursive Iteration

def printNums(num: Int): Unit = {
    if (num > 0) {
        print(s"$num ")
        printNums(num - 1)
    }
}
printNums(3)
println()

// For loop over collection

for (num <- list) {
    print(s"$num ")
}
println()

// for loop over multiple collections

for {
    num1 <- list
    num2 <- list
} print(s"$num1 $num2 ")
println()

// for with Option

val optionValue: Option[Int] = Option(33)
for (num <- optionValue) {
    println(s"$num")
}

// for with Either

def checkLowerLimit(value: Int): Either[String, Int] = {
    if (value > 20) Right(value)
    else Left("Invalid value")
}

def checkHigherLimit(value: Int): Either[String, String] = {
    if (value < 100) Right("Valid value")
    else Left("Invalid value")
}

println(for {
    value <- checkLowerLimit(55)
    value <- checkHigherLimit(value)
} yield value)
```
