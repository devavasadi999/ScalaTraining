### Operators
* Arithmetic
* Relational
* Logical
* Bitwise
* Assignment
* Unary
* Comparison and Reference Equality
* Operator Overloading
* Function Composition Operators

```scala
// Arithmetic

println(2 + 3) //addition
println(2 - 3) //subtraction
println(2 * 3) //multiplication
println(2 / 3) //division
println(2 % 3) //Modulus

// Relational

println(2 == 3) //equal to
println(2 != 3) // Not equal to
println(2 < 3) // less than
println(2 > 3) // greater than
println(2 >= 3) // greater than or equal to
println(2 <= 3) // less than or equal to

// Logical

println(true && false) // Logical AND
println(true || false) // Logical OR
println(!true) // Logical NOT

// Bitwise

println(2 & 3) // Bitwise AND
println(2 | 3) // Bitwise OR
println(2 ^ 3) // Bitwise XOR
println(~ 2) // Bitwise NOT
println(2 << 1) // Left Shift
println(-2 >> 1) // Right Shift with Sign Extension
println(2 >>> 1) // Right Shift with Zero Fill

// Assignment

var a = 1
a += 1 // addition assignment
a -= 1 // subtraction assignment
a *= 1 // multiplication assignment
a /= 1 // division assignment
a %= 1 // modulus assignment

// Unary

a = +3 // unary plus
a = -4 // unary minus
val b = !true // logical negation
a = ~a // bitwise negation

// Comparison and Reference Equality

val s1 = new String("Scala")
val s2 = new String("Scala")
println(s1 == s2) // value comparison
println(s1 eq s2) // reference comparison
println(s1 ne s2) // reference comparison

// Operator overloading -- Ex: a + b can be viewed as a.+(b)

class Point(val x: Int, val y: Int) {
    def +(p: Point): Point = new Point(x + p.x, y + p.y)
}

val p1 = new Point(1, 2)
val p2 = new Point(3, 4)
val p3 = p1 + p2
println(s"${p3.x} ${p3.y}")

// Function composition operators

// andThen -- execute first and pass the result to second

val add = (a:Int) => a + 1
val multiply = (a: Int) => a * 4

val addAndMultiply = add andThen multiply
println(addAndMultiply(2))

// compose -- execute second and pass the result to first

val multiplyandAdd = add compose multiply
println(multiplyandAdd(2))
```
